```python
f, ax = plt.subplots()

ax.bar(x=avg_monthly_precip.months,
       height=avg_monthly_precip.precip,
       color="cyan")

ax.set(title="Cyan plot of precip (in mm)")

plt.show()
```

```python
precip_by_seasons = avg_monthly_precip.groupby(["seasons"])[["precip"]].describe()
precip_by_seasons
```

```python
precip_by_seasons.columns
```

```python
precip_by_seasons.columns = precip_by_seasons.columns.droplevel(0)
precip_by_seasons
```

```python
f, ax = plt.subplots()

ax.bar(precip_by_seasons.index,
       precip_by_seasons["mean"],
       color="purple")

ax.set(title="something looks weird, but seasons")
plt.show()
```
